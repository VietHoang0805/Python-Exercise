```python
import matplotlib.pyplot as plt
import numpy as np
```

# Vẽ biểu đồ phương trình đa thức

```python
X = [-4,-3,-2,-1,0,1,2,3,4]
Y = [-4,-3,-2,-1,0,1,2,3,4]

curve = np.polyfit(X, Y, 2)
poly = np.poly1d(curve)

new_x = []
new_y = []
for i in range(-100,100):
    new_x.append(i+1)
    calc = poly(i+1)
    new_y.append(calc)

plt.plot(new_x, new_y)
plt.scatter(X,Y)
plt.show()
```
